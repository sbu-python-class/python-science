## `pivot_table` vs. `groupby`

see https://stackoverflow.com/questions/34702815/pandas-group-by-and-pivot-table-difference

```python
df = pd.DataFrame({"a": [1,2,3,1,2,3], "b":[1,1,1,2,2,2], "c":np.random.rand(6)})
df
```

A pivot table creates a new dataframe, with the specified column in the original treated as the index, and any columns given as columns in the dataframe, with the values taken to be "c"

```python
df.pivot_table("c", index=["a"], columns=["b"]) 
```

```python
df.groupby(['a','b'])['c'].sum()
```
